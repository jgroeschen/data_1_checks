```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
df = pd.read_csv('assets\songs_normalize.csv', header=0)

df_seconds = df
df_seconds['duration_s'] = df_seconds['duration_ms'] / 1000
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 10)
ax.xaxis.set_major_locator(MultipleLocator(5))
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.set_xlim(1997,2021)
plt.scatter(df_seconds['year'], df_seconds['duration_s'], )
plt.xlabel('Year', fontsize=12, labelpad=10)
plt.ylabel('Duration (seconds)', fontsize=12)
plt.title('Popular songs\' duration by year')
plt.annotate('Data source: Mark Koverha. 2022 May. Top Hits Spotify from 2000-2019, Version 1.\nRetrieved 2022-06-14 from https://www.kaggle.com/datasets/paradisejoy/top-hits-spotify-from-20002019',
             xy=(2000,100), xytext=(2021, 40), ha='right', bbox=dict(boxstyle='round', fc='white', alpha=0.5))
```

```python
explicit_by_year = df[['year', 'explicit']].groupby('year').mean()
explicit_by_year['explicit'] = explicit_by_year['explicit'] * 100
explicit_by_year.plot(kind='bar', legend=False, figsize=(15, 10), title='Percentage of top songs that were explicit, by year of release')
plt.xlabel('Year', fontsize=15, labelpad=10)
plt.ylabel('Percent Explicit', fontsize=15)
plt.grid(axis='y', linestyle='--')
plt.rc('axes', axisbelow=True)
plt.annotate('*Outliers - 1998 and 2020  only have 1 and 3 releases, respectively', xy=(0,0), xytext=(-2, -13),)
plt.annotate('Data source: Mark Koverha. 2022 May. Top Hits Spotify from 2000-2019, Version 1.\nRetrieved 2022-06-14 from https://www.kaggle.com/datasets/paradisejoy/top-hits-spotify-from-20002019',
             xy=(0,0), xytext=(23, -18), ha='right', bbox=dict(boxstyle='round', fc='white', alpha=0.5))

```
